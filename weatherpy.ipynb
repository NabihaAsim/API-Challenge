```python
import random 
import requests
import pandas as pd
from citipy import citipy 
import json
import seaborn as sns # seaborn library to set the aesthetics for the graphs
import matplotlib.pyplot as plt # for creating/plotting graphs
from datetime import datetime
```

```python

cities = []

while len(cities)<600:
    x = (random.choice(range(-90,90))) # randomly generate numbers in range -90 & 90 for latitude
    y = (random.choice(range(-180,180))) # randomly generagte numbers in range -180 & 180 for longitude
    city = citipy.nearest_city(x,y).city_name # look up city to match the lat & long randomly generated
    if city not in cities:
        cities.append(city)
```

```python
len(cities)
```

```python
url = "http://api.openweathermap.org/data/2.5/weather?"
api_key="546903a6c1f84e1bb7ff508d4a9479ce"

apiurl = url + "appid=" + api_key + "&q=" 

citywithdata = []   
latitude = []
longitude = []
temperature = []
humidity = []
cloudiness = []
wind_speed = []
dates = []

print(f'Beginning API Calls \n ........................')
for c in cities:
    city = c.replace(" ", "%20") 
    print(apiurl+city)
    response = requests.get(apiurl+city)
    response_dict = json.loads(response.text)
    if response.status_code != 404:
        citywithdata.append(city)
        temperature.append(response_dict['main']['temp_max'])
        latitude.append(response_dict['coord']['lat'])
        longitude.append(response_dict['coord']['lon'])
        humidity.append(response_dict['main']['humidity'])
        wind_speed.append(response_dict['wind']['speed'])
        dates.append(response_dict['dt'])
        cloudiness.append(response_dict['clouds']['all'])
print("End API Calls")
```

```python
weather_df = pd.DataFrame({"City": citywithdata,
                           "Latitude": latitude,
                           "Longitude": longitude,
                           "Humidity": humidity,
                           "Max Temp": temperature,
                           "Cloudiness": cloudiness,
                           "Wind Speed": wind_speed,
                           "Date": dates,
                          })

weather_df.head()
```

```python
weather_df.to_csv("WeatherPy.csv", encoding='utf-8', index=False)
```

```python
plt.figure(figsize=(10,8))
sns.set()
plt.scatter(weather_df["Latitude"], weather_df["Max Temp"], color='b', alpha=0.5)
plt.xlabel('Latitude')
plt.ylabel('Max Temperature (F)')
plt.title(f'City Latitude vs Max Temperature ({datetime.now().date()})')
plt.savefig("LatVsTemp.png")
plt.show()
```
